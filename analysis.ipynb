```python
import pandas as pd
from dataframe_info import DataFrameInfo

# Load the transformed data
df_transformed = pd.read_csv('/Users/mrhoti/Loan-Portfolio-Analysis-1/loan_data_transformed.csv')

# Instantiate the DataFrameInfo class
df_info = DataFrameInfo(df_transformed)

# Describe all columns in the DataFrame to check their data types
print("Column Data Types:")
print(df_info.describe_columns())

# Extract statistical values: median, standard deviation, and mean from the columns and the DataFrame
print("\nStatistical Values:")
print(df_info.extract_statistics())

# Count distinct values in categorical columns
print("\nDistinct Values in Categorical Columns:")
print(df_info.count_distinct_values())

# Print out the shape of the DataFrame
print("\nShape of the DataFrame:")
print(df_info.dataframe_shape())

# Generate a count/percentage count of NULL values in each column
print("\nNULL Value Information:")
print(df_info.null_value_counts())

# Print summary statistics
print("\nSummary Statistics:")
print(df_info.summary_statistics())

# Get unique values in a specific column (example)
column_name = 'grade'
print(f"\nUnique Values in Column '{column_name}':")
print(df_info.get_column_unique_values(column_name))

# Get value counts for a specific column (example)
column_name = 'loan_status'
print(f"\nValue Counts for Column '{column_name}':")
print(df_info.get_column_value_counts(column_name))
```
