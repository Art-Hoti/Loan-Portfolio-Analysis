```python
# Calculate total loan amount and total recoveries
total_loan_amount = df['loan_amount'].sum()
total_recoveries = df['recoveries'].sum()

# Calculate the percentage of recovered loans
recovered_percentage = (total_recoveries / total_loan_amount) * 100
print(f"Percentage of loans recovered: {recovered_percentage:.2f}%")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('loan_data.csv')

# Filter for charged off loans
charged_off_loans = df[df['loan_status'] == 'Charged Off'].copy()

# Fill NaN values in the 'term' column with a default value
charged_off_loans.loc[:, 'term'] = charged_off_loans['term'].fillna('0 months')

# Calculate the remaining term in months
charged_off_loans.loc[:, 'term_months'] = charged_off_loans['term'].apply(lambda x: int(x.split()[0]) if isinstance(x, str) else 0)

# Verify the intermediate calculation
print(charged_off_loans[['term', 'term_months']].head())

# Calculate the remaining term for each charged off loan
charged_off_loans.loc[:, 'remaining_term'] = charged_off_loans['term_months'] - (charged_off_loans['total_rec_prncp'] / charged_off_loans['loan_amount']) * charged_off_loans['term_months']

# Verify the intermediate calculation
print(charged_off_loans[['remaining_term']].head())

# Calculate the projected interest for the remaining term
charged_off_loans.loc[:, 'projected_interest'] = charged_off_loans['remaining_term'] * charged_off_loans['int_rate'] / 100 * charged_off_loans['out_prncp'] / 12

# Verify the intermediate calculation
print(charged_off_loans[['projected_interest']].head())

# Calculate the total projected loss
charged_off_loans.loc[:, 'projected_loss'] = charged_off_loans['out_prncp'] + charged_off_loans['projected_interest']

# Sum the total projected loss
total_projected_loss = charged_off_loans['projected_loss'].sum()
print(f"Total Projected Loss: ${total_projected_loss:,.2f}")

# Visualize the projected loss over the remaining term
plt.figure(figsize=(10, 6))
plt.plot(charged_off_loans['remaining_term'], charged_off_loans['projected_loss'], 'o', markersize=4, alpha=0.6)
plt.title('Projected Loss Over Remaining Term of Charged Off Loans')
plt.xlabel('Remaining Term (months)')
plt.ylabel('Projected Loss ($)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Load the data
df = pd.read_csv('loan_data.csv')

# Check column names
print(df.columns)

# Identify customers who are currently behind on payments
# Assuming 'loan_status' column to identify late payments
behind_on_payments = df[df['loan_status'].str.contains('Late|Past Due', na=False)].copy()

# Check for 'term_months' and other columns
print(behind_on_payments.columns)

# Total number of customers
total_customers = df.shape[0]

# Number of customers behind on payments
num_behind_on_payments = behind_on_payments.shape[0]

# Percentage of customers behind on payments
percentage_behind_on_payments = (num_behind_on_payments / total_customers) * 100
print(f"Percentage of customers behind on payments: {percentage_behind_on_payments:.2f}%")

# Calculate potential loss if these customers were charged off
behind_on_payments['potential_loss'] = behind_on_payments['out_prncp'].fillna(0)

# Calculate total potential loss
total_potential_loss = behind_on_payments['potential_loss'].sum()
print(f"Total potential loss if these customers were charged off: ${total_potential_loss:,.2f}")

# If 'term_months' is missing, handle alternative calculations
if 'term_months' in behind_on_payments.columns:
    # Calculate projected loss if customers finish the full loan term
    behind_on_payments['remaining_term'] = behind_on_payments['term_months'] - (behind_on_payments['total_rec_prncp'] / behind_on_payments['loan_amount']) * behind_on_payments['term_months']
    behind_on_payments['projected_interest'] = behind_on_payments['remaining_term'] * behind_on_payments['int_rate'].fillna(0) / 100 * behind_on_payments['loan_amount'] / 12
    behind_on_payments['projected_loss'] = behind_on_payments['potential_loss'] + behind_on_payments['projected_interest']

    # Total projected loss if they finish the full term
    total_projected_loss = behind_on_payments['projected_loss'].sum()
    print(f"Total projected loss if these customers finish the full loan term: ${total_projected_loss:,.2f}")
else:
    print("Column 'term_months' not found. Skipping projected loss calculation.")

# Calculate total expected revenue
total_expected_revenue = df['loan_amount'].sum()

# Revenue from late and charged off customers
charged_off_customers = df[df['loan_status'] == 'Charged Off']
revenue_late_and_defaulted = behind_on_payments['loan_amount'].sum() + charged_off_customers['loan_amount'].sum()

# Percentage of total expected revenue
percentage_revenue = (revenue_late_and_defaulted / total_expected_revenue) * 100
print(f"Percentage of total expected revenue represented by late and charged off customers: {percentage_revenue:.2f}%")
```
